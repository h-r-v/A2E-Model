```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import tensorflow as tf
seed = 1
```

```python
datafile = os.path.join( os.getcwd(), "Clean Data")
datafile = os.path.join( datafile, "data.csv")

df = pd.read_csv(datafile)
```

```python
df.shape, df.label.nunique()
```

```python
# TRAIN-VAL-TEST
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.1, stratify=df.label)
train, validate = train_test_split(train, test_size=0.11, stratify=train.label)

train.shape, validate.shape, test.shape
```

```python
labels = sorted(list(train.label.unique()))

import seaborn as sns
sns.countplot(x='label',data=train, order=labels)
```

```python
sns.countplot(x='label', data=validate, order=labels)
```

```python
sns.countplot(x='label', data=test, order=labels)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#AUGMENTATION

batch_size=16
target_size=(100, 100)
color_mode="rgb"

train_datagen = ImageDataGenerator(
        rescale=1./255,
        zoom_range=0.1,
        horizontal_flip=True,
        rotation_range=10,
        width_shift_range = 0.1,
        height_shift_range = 0.1,
        brightness_range = (0.5,1)
)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train,
    directory=None,
    x_col="path",
    y_col="label",
    target_size=target_size,
    color_mode=color_mode,
    class_mode="categorical",
    batch_size=batch_size,
    #save_to_dir = "preview",
    #save_prefix = "preview",
    #save_format = "png"
)

test_datagen = ImageDataGenerator(rescale=1./255)

validation_generator = test_datagen.flow_from_dataframe(
    dataframe=validate,
    directory=None,
    x_col="path",
    y_col="label",
    target_size=target_size,
    color_mode=color_mode,
    class_mode="categorical",
    batch_size=batch_size,
)

test_generator = test_datagen.flow_from_dataframe(
    dataframe=test,
    directory=None,
    x_col="path",
    y_col="label",
    target_size=target_size,
    color_mode=color_mode,
    class_mode="categorical",
    batch_size=batch_size,
    shuffle=False
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense
```

```python
#SHALLOW CNN
def create_cnn():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), input_shape=(100, 100, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(64, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(64, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())
    model.add(Dense(64))
    model.add(Activation('relu'))
    model.add(Dense(26))
    model.add(Activation('softmax'))

    model.compile(
                  loss='categorical_crossentropy', #becase this is not a multilabel dataset.
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.001, amsgrad=True),
                  metrics=['acc','categorical_accuracy']
                )
    
    model.summary()
    
    return model
```

```python
#CUSTOM GRID SEARCH
'''
- dropout
'''
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor = 'val_acc', mode = 'max', patience=25, restore_best_weights=True)
```

```python
model = create_cnn()
model_hist = model.fit( train_generator, 
                                 steps_per_epoch=train.shape[0] // batch_size, 
                                 epochs=150, 
                                 validation_data=validation_generator, 
                                 validation_steps=validate.shape[0] // batch_size,
                                 callbacks=[early_stop]
                      )
```

```python
plt.plot(model_hist.history['loss'], label = 'loss')
plt.plot(model_hist.history['val_loss'], label = 'val_loss')
plt.legend()
plt.show()

plt.plot(model_hist.history['acc'], label = 'acc')
plt.plot(model_hist.history['val_acc'], label = 'val_acc')
plt.legend()
plt.show()
```

```python
test_generator.reset()
test_acc = model.evaluate(test_generator)
```

```python
test_generator2 = test_datagen.flow_from_dataframe(
    dataframe=test,
    directory=None,
    x_col="path",
    y_col="label",
    target_size=target_size,
    color_mode=color_mode,
    class_mode="categorical",
    batch_size=1,
)

for i,idk in enumerate(test_generator2):
    
    plt.imshow(idk[0][0])
    plt.plot()
    
    ans = labels[np.argmax(idk[1])]
    prediction = labels[np.argmax(model.predict(idk[0]))]
    
    print('Ans: ', ans)
    print('Prediction: ', prediction)
    
    if i==0:
        break
```

```python
from sklearn.metrics import confusion_matrix

_ = lambda __: chr(__+ord('A'))

test_generator.reset()
guess = model.predict(test_generator)
guess = np.argmax(guess, axis=1)
ans = test_generator.classes

guess = list(map(_,guess))
ans = list(map(_,ans))

cm = confusion_matrix(ans,guess, labels=labels)

plt.figure(figsize = (8,8))
sns.heatmap(cm, xticklabels=labels, yticklabels=labels, annot=True)
plt.ylabel("True")
plt.xlabel("Predicted")
plt.plot()
```

```python
import datetime
now = datetime.datetime.now()
model_name = str(int(test_acc[1]*100)) + now.strftime("--%m-%d-%H-%M")+'.h5'

modeldir = os.path.join(os.getcwd(),'Models')
modelname = os.path.join(modeldir,model_name)

if os.path.exists(modeldir):
    model.save(modelname)
else:
    os.makedirs(modeldir)
    model.save(modelname)
    
print(model_name,"saved")
```
