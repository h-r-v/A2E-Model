```python
import os
import cv2
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
rootdir = os.getcwd()

datadir = os.path.join( rootdir, "Data")
newdatadir = os.path.join( rootdir, "Clean Data")

os.chdir(datadir)

if not os.path.exists(newdatadir):
    os.makedirs(newdatadir)
```

```python
def img_to_csv( img, label, filepath):
    #create o/p path
    if not os.path.exists( filepath ):
        with open( filepath, 'ab') as f:
            d = img.shape[0] * img.shape[1] + 1
            d = np.array(range(d)).reshape(-1, 1).T
            np.savetxt(f, d, delimiter=",")
    
    #save
    img = np.append(int(ord(label)-ord('A')),img)    
    img_array  = img.flatten().reshape(-1, 1).T
    print(img_array)

    with open( filepath, 'ab') as f:
        np.savetxt(f, img_array, delimiter=",")
```

```python
subdirs = os.listdir()

i=0
for subdir in subdirs:
    
    os.chdir(datadir)
    
    if os.path.isdir(subdir):
        
        newcwd = os.path.join( datadir, subdir)
        os.chdir(newcwd)
        
        for pic in os.listdir():
            if random.random() < 0.1:
                
                i+=1
                print(i,pic,subdir)
                img = cv2.imread(pic,0)
                
                resizedimg = cv2.resize(img, (100,100))
                
                img_to_csv( resizedimg , subdir, os.path.join( newdatadir, "data.csv") )
                
                plt.imshow(resizedimg, cmap='gray')
                plt.show()
                
os.chdir(datadir)
```
